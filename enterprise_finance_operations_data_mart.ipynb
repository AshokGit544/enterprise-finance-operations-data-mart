```python
!pip -q install pandas numpy matplotlib

import warnings
warnings.filterwarnings("ignore")

import random
from datetime import datetime, timedelta
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

random.seed(42)
np.random.seed(42)

pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 30)

PROJECT_DIR = Path("enterprise_finance_operations_data_mart_output")
PROJECT_DIR.mkdir(exist_ok=True)

print("Project directory:", PROJECT_DIR.resolve())

# Master data
def make_vendors(n=95):
    vendor_bases = [
        "Alpha Grid Services", "Northwind Electric Supply", "GreenVolt Systems", "Prime Utility Parts",
        "TransEnergy Logistics", "Vertex Maintenance", "BluePeak Consulting", "SolarEdge Components",
        "Metro Industrial Tools", "Crescent Field Ops", "Pioneer Cable Works", "Helios Analytics",
        "Summit Procurement", "IronBridge Equipment", "EnerSys Contractors", "Utility Edge Networks"
    ]
    categories = ["Maintenance", "IT Services", "Electrical Components", "Logistics", "Consulting", "Field Services"]
    onboarding_status = ["Active", "Active", "Active", "Review", "Inactive"]
    rows = []

    for i in range(1, n + 1):
        rows.append({
            "vendor_id": f"V{i:04d}",
            "vendor_name": random.choice(vendor_bases) + f" {i}",
            "vendor_category": random.choice(categories),
            "vendor_state": random.choice(["NY", "CT", "ME", "MA", "PA", "NJ", "RI"]),
            "payment_terms_days": random.choice([15, 30, 45, 60]),
            "preferred_vendor_flag": random.choice([1, 1, 1, 0]),
            "vendor_status": random.choice(onboarding_status)
        })
    return pd.DataFrame(rows)

def make_cost_centers():
    rows = [
        ("CC100", "Transmission Operations", "Operations"),
        ("CC110", "Distribution Maintenance", "Operations"),
        ("CC120", "Substation Engineering", "Engineering"),
        ("CC130", "Renewables Program", "Strategy"),
        ("CC140", "Customer Support", "Customer Ops"),
        ("CC150", "IT Shared Services", "Technology"),
        ("CC160", "Finance Operations", "Finance"),
        ("CC170", "Vegetation Management", "Field Ops"),
        ("CC180", "Grid Modernization", "Engineering"),
        ("CC190", "Regulatory Affairs", "Corporate")
    ]
    return pd.DataFrame(rows, columns=["cost_center_id", "cost_center_name", "department"])

def make_profit_centers():
    rows = [
        ("PC200", "Electric Distribution East", "East"),
        ("PC210", "Electric Distribution West", "West"),
        ("PC220", "Gas and Field Services", "North"),
        ("PC230", "Renewables and Storage", "Enterprise"),
        ("PC240", "Corporate Shared Services", "Corporate")
    ]
    return pd.DataFrame(rows, columns=["profit_center_id", "profit_center_name", "region_group"])

def make_gl_accounts():
    rows = [
        ("500100", "Opex - Maintenance", "Expense"),
        ("500200", "Opex - IT Services", "Expense"),
        ("500300", "Opex - Contractors", "Expense"),
        ("500400", "Opex - Logistics", "Expense"),
        ("500500", "Opex - Vegetation Mgmt", "Expense"),
        ("500600", "Opex - Consulting", "Expense"),
        ("500700", "Opex - Emergency Repairs", "Expense"),
        ("500800", "Capex - Grid Modernization", "Capital"),
        ("500900", "Capex - Substation Upgrades", "Capital"),
        ("110100", "Accounts Payable", "Liability")
    ]
    return pd.DataFrame(rows, columns=["gl_account", "gl_account_name", "gl_type"])

vendors = make_vendors()
cost_centers = make_cost_centers()
profit_centers = make_profit_centers()
gl_accounts = make_gl_accounts()

vendors.to_csv(PROJECT_DIR / "vendors.csv", index=False)
cost_centers.to_csv(PROJECT_DIR / "cost_centers.csv", index=False)
profit_centers.to_csv(PROJECT_DIR / "profit_centers.csv", index=False)
gl_accounts.to_csv(PROJECT_DIR / "gl_accounts.csv", index=False)

print("Master data created.")
print(vendors.shape, cost_centers.shape, profit_centers.shape, gl_accounts.shape)

# Source transactions
descriptions = [
    "substation maintenance support",
    "vegetation clearing around power lines",
    "grid inspection contractor services",
    "cybersecurity monitoring support",
    "cloud cost optimization consulting",
    "field equipment transportation",
    "metering systems enhancement",
    "SCADA support services",
    "storm response contractor mobilization",
    "distribution pole inspection",
    "asset reliability assessment",
    "emergency transformer replacement",
    "temporary field crew support",
    "circuit restoration activities",
    "renewables integration support",
    "substation modernization project"
]

invoice_statuses = ["RECEIVED", "VALIDATED", "APPROVED", "POSTED", "PAID", "HOLD"]

def generate_source_transactions(num_rows=6500):
    start_date = datetime(2024, 1, 1)
    end_date = datetime(2025, 12, 31)
    date_range_days = (end_date - start_date).days
    rows = []

    for i in range(1, num_rows + 1):
        vendor = vendors.sample(1).iloc[0]
        cc = cost_centers.sample(1).iloc[0]
        pc = profit_centers.sample(1).iloc[0]
        gl = gl_accounts.sample(1).iloc[0]

        invoice_received_date = start_date + timedelta(days=random.randint(0, date_range_days))
        validation_date = invoice_received_date + timedelta(days=random.randint(0, 6))
        approval_date = validation_date + timedelta(days=random.randint(0, 8))
        posting_date = approval_date + timedelta(days=random.randint(0, 5))
        due_date = posting_date + timedelta(days=int(vendor["payment_terms_days"]))
        payment_date = due_date + timedelta(days=random.randint(-4, 18))

        base_amount = random.uniform(2500, 95000)

        if vendor["vendor_category"] == "Consulting":
            base_amount *= 1.18
        if cc["cost_center_name"] == "Vegetation Management":
            base_amount *= 1.12
        if gl["gl_type"] == "Capital":
            base_amount *= 1.25
        if vendor["preferred_vendor_flag"] == 0:
            base_amount *= 1.05

        amount_usd = round(base_amount, 2)

        rows.append({
            "document_id": f"DOC{i:06d}",
            "invoice_id": f"INV{i:06d}",
            "vendor_id": vendor["vendor_id"],
            "cost_center_id": cc["cost_center_id"],
            "profit_center_id": pc["profit_center_id"],
            "gl_account": gl["gl_account"],
            "invoice_received_date": invoice_received_date.date().isoformat(),
            "validation_date": validation_date.date().isoformat(),
            "approval_date": approval_date.date().isoformat(),
            "posting_date": posting_date.date().isoformat(),
            "due_date": due_date.date().isoformat(),
            "payment_date": payment_date.date().isoformat(),
            "amount_usd": amount_usd,
            "currency": "USD",
            "description": random.choice(descriptions),
            "status": random.choice(invoice_statuses),
            "source_system": random.choice(["SAP_FI", "SAP_AP", "ERP_STAGE"]),
            "payment_method": random.choice(["ACH", "Wire", "Check"]),
            "invoice_type": random.choice(["Standard", "Service", "Emergency", "Capital"])
        })

    return pd.DataFrame(rows)

source_transactions = generate_source_transactions()

# Inject data issues and operational exceptions like real enterprise feeds
issue_idx = np.random.choice(source_transactions.index, size=260, replace=False)

missing_vendor_idx = issue_idx[:40]
source_transactions.loc[missing_vendor_idx, "vendor_id"] = None

missing_gl_idx = issue_idx[40:70]
source_transactions.loc[missing_gl_idx, "gl_account"] = None

hold_idx = issue_idx[70:130]
source_transactions.loc[hold_idx, "status"] = "HOLD"

late_validation_idx = issue_idx[130:180]
source_transactions.loc[late_validation_idx, "validation_date"] = (
    pd.to_datetime(source_transactions.loc[late_validation_idx, "invoice_received_date"]) +
    pd.to_timedelta(np.random.randint(7, 21, size=len(late_validation_idx)), unit="D")
).dt.date.astype(str)

late_posting_idx = issue_idx[180:220]
source_transactions.loc[late_posting_idx, "posting_date"] = (
    pd.to_datetime(source_transactions.loc[late_posting_idx, "approval_date"]) +
    pd.to_timedelta(np.random.randint(6, 18, size=len(late_posting_idx)), unit="D")
).dt.date.astype(str)

overdue_payment_idx = issue_idx[220:]
source_transactions.loc[overdue_payment_idx, "payment_date"] = (
    pd.to_datetime(source_transactions.loc[overdue_payment_idx, "due_date"]) +
    pd.to_timedelta(np.random.randint(20, 75, size=len(overdue_payment_idx)), unit="D")
).dt.date.astype(str)

source_transactions.to_csv(PROJECT_DIR / "source_transactions.csv", index=False)
print("Source transactions created:", source_transactions.shape)
display(source_transactions.head())

# Bronze layer
bronze_df = source_transactions.copy()
bronze_df.to_csv(PROJECT_DIR / "bronze_transactions.csv", index=False)

# Silver layer: standardize and enrich
silver_df = bronze_df.copy()

date_cols = [
    "invoice_received_date", "validation_date", "approval_date",
    "posting_date", "due_date", "payment_date"
]
for c in date_cols:
    silver_df[c] = pd.to_datetime(silver_df[c], errors="coerce")

silver_df = (
    silver_df
    .merge(vendors, on="vendor_id", how="left")
    .merge(cost_centers, on="cost_center_id", how="left")
    .merge(profit_centers, on="profit_center_id", how="left")
    .merge(gl_accounts, on="gl_account", how="left")
)

silver_df["invoice_month"] = silver_df["invoice_received_date"].dt.to_period("M").astype(str)
silver_df["posting_month"] = silver_df["posting_date"].dt.to_period("M").astype(str)

silver_df["validation_sla_days"] = (silver_df["validation_date"] - silver_df["invoice_received_date"]).dt.days
silver_df["approval_sla_days"] = (silver_df["approval_date"] - silver_df["validation_date"]).dt.days
silver_df["posting_sla_days"] = (silver_df["posting_date"] - silver_df["approval_date"]).dt.days
silver_df["payment_lag_days"] = (silver_df["payment_date"] - silver_df["due_date"]).dt.days

silver_df["master_data_error_flag"] = (
    silver_df["vendor_name"].isna() |
    silver_df["gl_account_name"].isna() |
    silver_df["cost_center_name"].isna() |
    silver_df["profit_center_name"].isna()
).astype(int)

silver_df["validation_sla_breach_flag"] = (silver_df["validation_sla_days"] > 5).astype(int)
silver_df["approval_sla_breach_flag"] = (silver_df["approval_sla_days"] > 5).astype(int)
silver_df["posting_sla_breach_flag"] = (silver_df["posting_sla_days"] > 3).astype(int)
silver_df["payment_overdue_flag"] = (silver_df["payment_lag_days"] > 0).astype(int)
silver_df["on_hold_flag"] = (silver_df["status"] == "HOLD").astype(int)

silver_df["exception_reason"] = ""

silver_df.loc[silver_df["master_data_error_flag"] == 1, "exception_reason"] += "Missing master data | "
silver_df.loc[silver_df["validation_sla_breach_flag"] == 1, "exception_reason"] += "Validation SLA breach | "
silver_df.loc[silver_df["approval_sla_breach_flag"] == 1, "exception_reason"] += "Approval SLA breach | "
silver_df.loc[silver_df["posting_sla_breach_flag"] == 1, "exception_reason"] += "Posting SLA breach | "
silver_df.loc[silver_df["payment_overdue_flag"] == 1, "exception_reason"] += "Payment overdue | "
silver_df.loc[silver_df["on_hold_flag"] == 1, "exception_reason"] += "Invoice on hold | "

silver_df["exception_reason"] = silver_df["exception_reason"].str.rstrip(" |")
silver_df.to_csv(PROJECT_DIR / "silver_enriched_transactions.csv", index=False)

print("Silver layer ready:", silver_df.shape)
display(silver_df.head())

# Gold layer: KPI mart
gold_kpi_mart = (
    silver_df.groupby(["posting_month", "department", "gl_type"], as_index=False)
    .agg(
        transaction_count=("document_id", "count"),
        total_amount_usd=("amount_usd", "sum"),
        avg_amount_usd=("amount_usd", "mean"),
        master_data_errors=("master_data_error_flag", "sum"),
        validation_sla_breaches=("validation_sla_breach_flag", "sum"),
        approval_sla_breaches=("approval_sla_breach_flag", "sum"),
        posting_sla_breaches=("posting_sla_breach_flag", "sum"),
        overdue_payments=("payment_overdue_flag", "sum"),
        hold_invoices=("on_hold_flag", "sum")
    )
    .sort_values(["posting_month", "department", "gl_type"])
)

gold_kpi_mart["master_data_error_pct"] = (
    gold_kpi_mart["master_data_errors"] / gold_kpi_mart["transaction_count"] * 100
).round(2)
gold_kpi_mart["validation_sla_breach_pct"] = (
    gold_kpi_mart["validation_sla_breaches"] / gold_kpi_mart["transaction_count"] * 100
).round(2)
gold_kpi_mart["approval_sla_breach_pct"] = (
    gold_kpi_mart["approval_sla_breaches"] / gold_kpi_mart["transaction_count"] * 100
).round(2)
gold_kpi_mart["posting_sla_breach_pct"] = (
    gold_kpi_mart["posting_sla_breaches"] / gold_kpi_mart["transaction_count"] * 100
).round(2)
gold_kpi_mart["overdue_payment_pct"] = (
    gold_kpi_mart["overdue_payments"] / gold_kpi_mart["transaction_count"] * 100
).round(2)
gold_kpi_mart["hold_invoice_pct"] = (
    gold_kpi_mart["hold_invoices"] / gold_kpi_mart["transaction_count"] * 100
).round(2)

gold_kpi_mart.to_csv(PROJECT_DIR / "gold_finance_kpi_mart.csv", index=False)

# Exception queue
exception_queue = silver_df[
    (silver_df["master_data_error_flag"] == 1) |
    (silver_df["validation_sla_breach_flag"] == 1) |
    (silver_df["approval_sla_breach_flag"] == 1) |
    (silver_df["posting_sla_breach_flag"] == 1) |
    (silver_df["payment_overdue_flag"] == 1) |
    (silver_df["on_hold_flag"] == 1)
].copy()

exception_queue["exception_count"] = (
    exception_queue["master_data_error_flag"] +
    exception_queue["validation_sla_breach_flag"] +
    exception_queue["approval_sla_breach_flag"] +
    exception_queue["posting_sla_breach_flag"] +
    exception_queue["payment_overdue_flag"] +
    exception_queue["on_hold_flag"]
)

exception_queue["priority"] = np.select(
    [
        exception_queue["exception_count"] >= 4,
        exception_queue["exception_count"] >= 2,
        exception_queue["exception_count"] == 1
    ],
    ["Critical", "High", "Medium"],
    default="Low"
)

exception_queue = exception_queue.sort_values(
    ["exception_count", "amount_usd"], ascending=[False, False]
)

exception_queue.to_csv(PROJECT_DIR / "finance_exception_queue.csv", index=False)

# Operational SLA summary
sla_summary = pd.DataFrame({
    "metric": [
        "Validation SLA <= 5 days",
        "Approval SLA <= 5 days",
        "Posting SLA <= 3 days",
        "Payment overdue rate",
        "Master data error rate",
        "On-hold invoice rate"
    ],
    "value_pct": [
        round((1 - silver_df["validation_sla_breach_flag"].mean()) * 100, 2),
        round((1 - silver_df["approval_sla_breach_flag"].mean()) * 100, 2),
        round((1 - silver_df["posting_sla_breach_flag"].mean()) * 100, 2),
        round(silver_df["payment_overdue_flag"].mean() * 100, 2),
        round(silver_df["master_data_error_flag"].mean() * 100, 2),
        round(silver_df["on_hold_flag"].mean() * 100, 2)
    ]
})
sla_summary.to_csv(PROJECT_DIR / "sla_summary.csv", index=False)

# Vendor operations scorecard
vendor_scorecard = (
    silver_df.groupby(["vendor_id", "vendor_name", "vendor_category", "vendor_status"], as_index=False)
    .agg(
        transaction_count=("document_id", "count"),
        total_amount_usd=("amount_usd", "sum"),
        overdue_payments=("payment_overdue_flag", "sum"),
        hold_invoices=("on_hold_flag", "sum"),
        master_data_errors=("master_data_error_flag", "sum"),
        validation_sla_breaches=("validation_sla_breach_flag", "sum")
    )
)

vendor_scorecard["operational_issue_count"] = (
    vendor_scorecard["overdue_payments"] +
    vendor_scorecard["hold_invoices"] +
    vendor_scorecard["master_data_errors"] +
    vendor_scorecard["validation_sla_breaches"]
)

vendor_scorecard["operational_issue_pct"] = (
    vendor_scorecard["operational_issue_count"] / vendor_scorecard["transaction_count"] * 100
).round(2)

vendor_scorecard = vendor_scorecard.sort_values(
    ["operational_issue_pct", "total_amount_usd"], ascending=[False, False]
)
vendor_scorecard.to_csv(PROJECT_DIR / "vendor_operations_scorecard.csv", index=False)

# Department operations summary
department_ops_summary = (
    silver_df.groupby(["department"], as_index=False)
    .agg(
        transaction_count=("document_id", "count"),
        total_amount_usd=("amount_usd", "sum"),
        overdue_payments=("payment_overdue_flag", "sum"),
        holds=("on_hold_flag", "sum"),
        validation_breaches=("validation_sla_breach_flag", "sum"),
        posting_breaches=("posting_sla_breach_flag", "sum")
    )
)

department_ops_summary["overall_issue_pct"] = (
    (department_ops_summary["overdue_payments"] +
     department_ops_summary["holds"] +
     department_ops_summary["validation_breaches"] +
     department_ops_summary["posting_breaches"]) /
    department_ops_summary["transaction_count"] * 100
).round(2)

department_ops_summary = department_ops_summary.sort_values("overall_issue_pct", ascending=False)
department_ops_summary.to_csv(PROJECT_DIR / "department_operations_summary.csv", index=False)

# Finance operations daily queue
daily_queue = (
    silver_df.groupby(["posting_date"], as_index=False)
    .agg(
        transaction_count=("document_id", "count"),
        total_amount_usd=("amount_usd", "sum"),
        hold_count=("on_hold_flag", "sum"),
        overdue_count=("payment_overdue_flag", "sum"),
        master_data_errors=("master_data_error_flag", "sum")
    )
    .sort_values("posting_date")
)
daily_queue.to_csv(PROJECT_DIR / "daily_finance_operations_queue.csv", index=False)

# Simple charts
top_monthly_amount = (
    silver_df.groupby("posting_month", as_index=False)
    .agg(total_amount_usd=("amount_usd", "sum"))
    .sort_values("posting_month")
)

plt.figure(figsize=(12, 5))
plt.plot(top_monthly_amount["posting_month"], top_monthly_amount["total_amount_usd"], marker="o")
plt.xticks(rotation=45, ha="right")
plt.title("Monthly Finance Processing Amount")
plt.ylabel("Total Amount USD")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
plt.bar(department_ops_summary["department"], department_ops_summary["overall_issue_pct"])
plt.xticks(rotation=45, ha="right")
plt.title("Department Operational Issue Percentage")
plt.ylabel("Issue Percentage")
plt.tight_layout()
plt.show()

# Show main outputs
print("\nSLA Summary")
display(sla_summary)

print("\nTop exception queue items")
display(exception_queue[[
    "document_id", "invoice_id", "vendor_name", "department", "gl_account_name",
    "amount_usd", "status", "exception_count", "priority", "exception_reason"
]].head(20))

print("\nVendor operations scorecard")
display(vendor_scorecard.head(15))

print("\nGenerated files:")
for p in sorted(PROJECT_DIR.glob("*")):
    print("-", p.name)
```
